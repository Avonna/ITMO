```python
import pandas as pd
```

```python
data = pd.read_csv('/content/ITMOUniversity+DATST+spring_2021_ITMO_bac+type@asset+block@task-9cfcb652-7cc6-11eb-adfa-367dda533c29.csv', 
                       delimiter=',', 
                       decimal='.', index_col = 'ID')
```

Посмотрим на наши данные

```python
data
```

Теперь выполним экспоненциальную нормировку:
$$
\widetilde x_i = 1 - e^{1 - \frac{x_i}{x_{min}}}
$$

Подключим библиотеку для вычисления экспоненты

```python
import numpy as np
```

Проведем экспоненциальную нормировку и выведем данные

```python
data = (data - data.min())/data.max()-data.min()
data
```

```python
# copy the data
df_min_max_scaled = data.copy()

# apply normalization techniques
for column in df_min_max_scaled.columns:
	df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())	

# view normalized data
df_min_max_scaled
```

Заведем дополнительную колонку для вычисления значений целевой функции:

```python
df_min_max_scaled['SUM'] = 0
```

Вычислим значение целевой функции и отсортируем результат:

```python
df_min_max_scaled['SUM'] = df_min_max_scaled['GUESTS'] + df_min_max_scaled['INCOME'] + df_min_max_scaled['STARS']
df_min_max_scaled.sort_values(by = ['SUM'])
```
